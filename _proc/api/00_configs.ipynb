# Configurations

There are several layers that require & allow configuration of *findmycells*. However, only the top most level ([`ProjectConfigs`](https://Defense-Circuits-Lab.github.io/findmycells/api/configs.html#projectconfigs)) is fully implemented here. Both lower level configs (e.g. configs for each processing type like "preprocessing" or "quantification", or even lower level configs like for each individual processing step represented by [`ProcessingStrategy`](https://Defense-Circuits-Lab.github.io/findmycells/api/core.html#processingstrategy) subclasses), will be defined by each of these classes using the [`DefaultConfigs`](https://Defense-Circuits-Lab.github.io/findmycells/api/configs.html#defaultconfigs) - which, inturn, is implemented here.
